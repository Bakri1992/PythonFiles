```python
class Order:
    def __init__(self,cart,customer):
        self.cart=cart
        self.customer=customer
    # __len__ describes my len() function
    #make sure that returns integer not float or anything else
    def __len__(self):
        return len(self.cart) # Number of items   
    
    # To specify what happens when we call object() => __call__
    def __call__(self):
        print(f"{self.customer}")
        
    # To specify what happens when we print(object) => __str__
    # make sure that returns string not float or anything else
    def __str__(self):
        return f"'{self.customer}' bought {self.cart} "
    # Normally we use repr to represent the object
    # It replace __str__ when its not exists
    # we I have both and print(object) __str__ will be executed
    def __repr__(self):
        return f" {self.__class__.__name__}(cart= {self.cart} , customer =({self.customer}) )"
    
    # To specify if my object True or False => __bool__
    def __bool__(self):
        return len(self.cart)>0
    
    # To add item to my cart => __add__ it represents my + operator
    def __add__(self,item):
        self.cart.append(item)
        return self
        
    # To use += operator => __iadd__  
    def __iadd__ (self,item):
        self.cart.append(item)
        return self
    # * => __mul__    *=  => __imul__
    # / => __div__    /=  => __idiv__
    # - => __sub__    -=  => __isub__
    
    # what if I want to add my item at the beginning and I added object to the item
    # we use  __radd__
    def __radd__(self,item):
        self.cart.insert(0,item)
        return self
    
    
    # How to set user defined indexing => __getitem__
    
    def __getitem__(self,key):
        return self.cart[key]
    
    
    # How to set an item using an index => __setitem__
    
    def __setitem__(self,key,value):
        self.cart[key]=value
        return self
        
```

```python
order1 =Order(["Laptop","Monitor","Mouse","Keyboard"],"Bakri Badet")
print(len(order1))
print(order1)
print(repr(order1))
if order1: # here the bool(order1) will be called
    print(f"{order1.customer} order is processing")
else: 
    print(f"Shooping cart is empty")
    
# Lets add a new item to my order
order1+"Mousepad"
print(order1.cart)

# Lets add item using +=
order1 += "USB Stick"
# Here __add__ will be used
print(order1.cart)


# Lets add the item at the beginning
"Case"+ order1
# Here __radd__ will be used
print(order1.cart)

print(order1[1])

order1[1]="Pencil"

print(order1.cart)
```
