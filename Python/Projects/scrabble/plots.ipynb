```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ergebnisse nach erstem Durchlauf f√ºr Set und Liste
# mit Filterung der Wortliste nach max. 7 Buchstaben
res = pd.DataFrame({"List":[0.08,178.8,809.92,3570.05,16300.54,None,None],"Set":[4772.85,469.4,470.33,538.26,394.9,608.33,1151.08]})
```

```python
plt.plot(range(1,8),res.Set, color="darkblue", label="set")
plt.plot(range(1,8),res.List, color="darkred", label="list")
plt.ylabel("seconds")
plt.xlabel("length of input")
plt.legend()
plt.show()
```

```python
res_binary = res.copy()
res_binary["list_binary"] = [0.04,2.04,10.21,37.13,185.94,1396.12,8102.49]
```

```python
plt.plot(range(1,8),res_binary.Set, color="darkblue", label="set")
plt.plot(range(1,8),res_binary.List, color="darkred", label="list")
plt.plot(range(1,8),res_binary.list_binary, color="red", label="list_binary")
plt.ylabel("seconds")
plt.xlabel("length of input")
plt.legend()
plt.show()
```

```python
res_binary2 = res_binary.copy()
res_binary2["list_binary2"] = [0.04,2.46,13.13,58.72,294.04,1367.49,6978.97]
```

```python
plt.plot(range(1,8),res_binary2.Set, color="darkblue", label="set")
plt.plot(range(1,8),res_binary2.List, color="darkred", label="list")
plt.plot(range(1,8),res_binary2.list_binary, color="red", label="list_binary")
plt.plot(range(1,8),res_binary2.list_binary2, color="pink", label="list_binary ohne Filter")
plt.ylabel("seconds")
plt.xlabel("length of input")
plt.legend()
plt.show()
```
